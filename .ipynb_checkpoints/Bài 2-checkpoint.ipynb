```python
def get_len(word):
    try:
        n = len(word.split())
    except:
        return 0
    else:
        return n
```

```python
"""
thông tin sinh viên có câu trả lời
dài nhất cho câu hỏi "Theo bạn, nhà trường và giáo viên cần phải làm gì để khắc phục
tình trạng lười học ở sinh viên?"
"""

counts = df['B2'].apply(get_len)
df.iloc[counts.argmax()][['Họ và tên', 'Giới tính', 'Chuyên ngành']]
```

```python
import re
```

```python
def get_major(major):
    major = major.lower()
    if 'cntt' in major:
        return 'cntt'.upper()
    else:
        search = re.search(r'(.+)-(.+)', major)
        if search != None:
            return search.group(1).upper()
        else:
            return major.upper()
```

```python
"""
Các chuyên ngành được hỏi nhiều nhất
"""

mj = df['Chuyên ngành'].map(get_major)
mj.value_counts().head()
```

```python
mj.value_counts().head().plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon', 'gold', 'orange'])
plt.ylabel('Số lượng')
plt.title('Các chuyên ngành được hỏi nhiều nhất')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Trực quan mức độ hài lòng của các câu hỏi

```python
colors = ['skyblue', 'lightgreen', 'salmon', 'gold', 'orange']

fig, axes = plt.subplots(3, 3, figsize=(20, 15))

axes = axes.flatten()

for i, column in enumerate(df.select_dtypes('number')):
    ax = axes[i]
    n = df[column].nunique()
    ax.bar(df[column].value_counts().index, df[column].value_counts(), color=colors[:n])
    ax.set_title('Câu hỏi số {}'.format(column[-1]))
    ax.set_xlabel('Mức độ hài lòng')
    ax.set_ylabel('Số lượng')
    ax.tick_params(axis='x')

fig.suptitle('Mức độ hài lòng của từng câu hỏi khảo sát', fontsize=16)
plt.tight_layout()

plt.show()
```

```python
from underthesea import word_tokenize
from underthesea import sent_tokenize
```

```python
def preprocessing(data):
    split_data = []
    for x in data:
        try:
            temp = sent_tokenize(x)
        except:
            split_data.append('None')
        else:
            split_data.extend(temp)
    for i in range(len(split_data)):
        split_data[i] = re.sub('[^\w\s]', '', split_data[i])
    data = []
    for x in split_data:
        data.extend(word_tokenize(x))
    return data
```

```python
colors = ['skyblue', 'lightgreen', 'salmon', 'gold', 'orange', 'purple', 'cyan', 'pink']
```

```python
def visualize(data, title): 
    n = data.nunique()
    data.value_counts().plot.bar(figsize=(12, 6), color=colors[:n])
    plt.title(title)
    plt.xlabel('Phản hồi')
    plt.ylabel('Số lần xuất hiện')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
def filters(word):
    word = word.lower()
    if 'lười' in word:
        return 'lười'
    elif 'chăm' in word:
        return 'chăm'
    elif 'bình thường' in word:
        return 'bình thường'
    else:
        return 'ý kiến khác'
```

```python
data = df['B1'].map(filters)
```

```python
visualize(pd.Series(data), 'Tần suất của các phản hồi về việc là người chăm học hay lười học')
```

```python
def filters(word):
    try:
        word = word.lower()
    except:
        return 'ý kiến khác'
    else:
        x1 = list('cải thiện,tăng cường,chú trọng,đổi mới,tích cực,chất lượng,thay đổi,tương tác,phương pháp'.split(','))
        x2 = list('khích lệ,khuyến khích,động viên,khích lệ,quan tâm,tuyên truyền,động viên,nhắc nhở'.split())
        x3 = list('ý thức,bản thân'.split())

        for x in x1:
            if x in word:
                return 'cải thiện chất lượng'
        for x in x2:
            if x in word:
                return 'quan tâm sinh viên'
        for x in x3:
            if x in word:
                return 'ý thức sinh viên'
        return 'ý kiến khác'
```

```python
data = df['B2'].map(filters)
visualize(pd.Series(data), 'Theo bạn, nhà trường và giáo viên cần phải làm gì để khắc phục tình trạng lười học ở sinh viên?')
```
